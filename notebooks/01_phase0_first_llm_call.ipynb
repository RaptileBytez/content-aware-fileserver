# Phase 0: First LLM Call

Goal of this notebook is testing the locally installed LLM

 `bartowski/allura-forge_llama-3.3-8B-Instruct-GGUF` 
 
 a 8Bit quantified LLama 3.3 model.
The test will be to send a system instruction and a prompt to create a German lyric in the style of [Johann Wolfgang von Goethe](https://en.wikipedia.org/wiki/Johann_Wolfgang_von_Goethe) 

```python
from openai import OpenAI
from dotenv import load_dotenv
import os

# Configuration
load_dotenv()
client = OpenAI(api_key=os.getenv("LM_STUDIO_API_KEY"), base_url=os.getenv("LM_STUDIO_BASE_URL"))
```

```python
# Send a prompt
model_name = os.getenv("LM_STUDIO_MODEL")
if not model_name:
    raise ValueError("LM_STUDIO_MODEL not configured. Check your .env file.")

completion = client.chat.completions.create(
    model=model_name,  # exakter Name aus LM Studio
    messages=[
        {"role": "system", "content": "Du bist Johann Wolfgang von Goethe."},
        {"role": "user", "content": "Schreibe ein Gedicht über die Programmiersprache Python."}
    ]
)
```

```python
# Output
print(completion.choices[0].message.content)
```

## Conclusion

The following key points can be taken from Phase 0 of the project:

### LM Studio as a Local LLM Server
The API provided by LM Studio is an OpenAI-compatible REST API. This allows code written for the industry standard OpenAI API to run on a local model with only minor modifications. As shown, only the `base_url` and the `model` need to be adjusted.

### Configuration via Environment Variables
API_KEY, server URL, and model name are managed via a `.env` file. This separates logic from configuration and follows best practices.

---

### Roles in the Chat Format
LLMs expect inputs in a structured format with `system` and `user` roles. The `system` role defines the behavior of the model, while the `user` role contains the actual prompt.
