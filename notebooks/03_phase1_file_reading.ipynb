```python
from pathlib import Path
from typing import Any
from pydantic import BaseModel
```

```python
class DocumentFile(BaseModel):
    model_config = {"arbitrary_types_allowed": True}
    file_name: str
    suffix: str
    path: Path
```

```python
def crawl_directory(path: Path, files: list[DocumentFile] | None = None) -> dict[str, Any]:
    if files is None:
        files = []
    for elem in path.iterdir():
        if elem.is_file():
            file = DocumentFile(file_name=elem.name, suffix=elem.suffix, path=elem)
            files.append(file)
        if elem.is_dir():
            crawl_directory(elem, files)
    content = {
        "files": files,
        "num_files": len(files)
    }
    return content
```

```python
data_dir = Path.cwd().parents[0] / "data"
d = crawl_directory(data_dir)
print(d)
```

## Supported File Types
PDF extraction is implemented first. The following types will be 
added in subsequent phases, each requiring a different extraction strategy:
- Images (.png, .jpg): require OCR
- Office documents (.docx, .xlsx): require dedicated parsers
- Emails (.eml, .msg): require email parsing
- E-invoices (.pdf with ZUGFeRD XML): structured data extraction
